```python
import pandas as pd
import numpy as np  
```

```python
#caracteristicas generales, seguridad social en salud y educación 

#links a los conjuntos de datos: 
url_cgsse = 'https://raw.githubusercontent.com/niconomist98/DataAnalyticsUQ/refs/heads/main/Datos/GEIH2025/cgsse.CSV'
url_ocupados = 'https://github.com/niconomist98/DataAnalyticsUQ/raw/refs/heads/main/Datos/GEIH2025/Ocupados.CSV'
url_ft = 'https://github.com/niconomist98/DataAnalyticsUQ/raw/refs/heads/main/Datos/GEIH2025/Fuerza%20de%20trabajo.CSV' 
url_desocupados = 'https://github.com/niconomist98/DataAnalyticsUQ/raw/refs/heads/main/Datos/GEIH2025/No%20ocupados.CSV'
```

```python
# CGSSE
df_cgsse = pd.read_csv(url_cgsse, encoding='latin-1', sep=';', 
                       low_memory=False)

#Ocupados
df_ocupados=pd.read_csv(url_ocupados,encoding='latin-1',sep=';',
                        low_memory=False)

#Fuerza de trabajo
df_ft=pd.read_csv(url_ft,
                  encoding='latin-1',sep=';',low_memory=False)
#No ocupados
df_desocupados=pd.read_csv(url_desocupados,encoding='latin-1',
                           sep=';',low_memory=False)
```

```python
poblacion_total = df_cgsse["FEX_C18"].sum()
print(f"esta es la poblacion total: {poblacion_total}")
```

```python
# codigo de cuantos años tiene es P6040
PET= df_cgsse[(df_cgsse["P6040"]>=15) & (df_cgsse["P6040"]<=65)]["FEX_C18"].sum()
print(f"esta es la poblacion en edad de trabajar es :{PET}")
```

```python
# codigo de cuantos años tiene es P6040
PENT= df_cgsse[df_cgsse["P6040"]<15]["FEX_C18"].sum()
PENT_2= df_cgsse[df_cgsse["P6040"]>65]["FEX_C18"].sum()
print(f"esta es la poblacion que no esta en edad de trabajar es :{PENT + PENT_2}")
```

```python
# pobalcion economicamente activa es variable "FT" en df_ft
# FT=1  si trabaja 
PEA= df_ft[df_ft["FT"]==1]["FEX_C18"].sum()
print(f"poblacion economicamente activa {PEA}")
```

```python
# pobalcion economicamente activa es variable "FT" en df_ft
# FFT  no trabaja 
PEI= df_ft[df_ft["FFT"]==1]["FEX_C18"].sum()
print(f"poblacion economicamente activa {PEI}")
```

```python
# OCI = 1 si ocupado
ocupados_pob= df_ocupados[(df_ocupados["OCI"]==1)]["FEX_C18"].sum()
print(f"poblavion ocupada {ocupados_pob}")
```

```python
# DSI : desocupado? si =1
desocupado_pob = df_desocupados[(df_desocupados["DSI"]==1)]["FEX_C18"].sum()
print(F" esta es la tasa de la poblacion desocupada {desocupado_pob}")
```

```python
# tasa de ocupacion 
tasa_ocupacion = ocupados_pob/PET
print(f"esta es la tasa de ocupacion {tasa_ocupacion}")
```

```python
# tasa de desempleo
tasa_de_desempleo = desocupado_pob/PEA
print(f"tasa de desempleo {tasa_de_desempleo*100}%")
```

```python
#Primero hay que anexar la p3271 (sexo), 1=H a 2=M o sea de sexos a la tabla de desocupados Y fuerza laboral
df_desocupados = pd.merge(df_desocupados, df_cgsse[['DIRECTORIO', 'SECUENCIA_P', 'ORDEN', 'P3271','P6040']], 
                        on=['DIRECTORIO', 'SECUENCIA_P', 'ORDEN'], how='left')

df_ft = pd.merge(df_ft, df_cgsse[['DIRECTORIO', 'SECUENCIA_P', 'ORDEN', 'P3271','P6040']], 
                        on=['DIRECTORIO', 'SECUENCIA_P', 'ORDEN'], how='left')
```

# tasas de desempleo hombres y mujeres tolima, caldas,risaralda y quindio

```python
# 1. Calcular cantidad de hombres desempleados en el quindío 
HDQ = df_desocupados[(df_desocupados['DSI']==1) & (df_desocupados['DPTO']==63) & (df_desocupados['P3271']==1)]['FEX_C18'].sum()
print(f'Hombres Desempleados en el Quindio: {HDQ}')

# 2. Calcular cantidad de hombres en la PEA del quindío
PEAHQ = df_ft[(df_ft['FT']==1) & (df_ft['DPTO']==63) & (df_ft['P3271']==1) ]['FEX_C18'].sum()
print(f'Poblacion Economicamente Activa de hombres en el quindío: {PEAHQ}')

# 3. Calcular la tasa de desempleo en hombres del quindío
TDHQ = HDQ/PEAHQ
print(f'Tasa de Desempleo en hombres del quindío: {TDHQ}')
```

```python
# 1. Calcular cantidad de mujeres desempleados en el quindío 
MDQ = df_desocupados[(df_desocupados['DSI']==1) & (df_desocupados['DPTO']==63) & (df_desocupados['P3271']==2)]['FEX_C18'].sum()
print(f'mujeres Desempleados en el Quindio: {MDQ}')

# 2. Calcular cantidad de mujeres en la PEA del quindío
PEAMQ = df_ft[(df_ft['FT']==1) & (df_ft['DPTO']==63) & (df_ft['P3271']==2) ]['FEX_C18'].sum()
print(f'Poblacion Economicamente Activa de mujeres en el quindío: {PEAMQ}')

# 3. Calcular la tasa de desempleo en mujeres del quindío
TDMQ = MDQ/PEAMQ
print(f'Tasa de Desempleo en mujeres del quindío: {TDMQ}')
```

```python
# 1. Calcular cantidad de hombres desempleados en risaralda
HDR = df_desocupados[(df_desocupados['DSI']==1) & (df_desocupados['DPTO']==66) & (df_desocupados['P3271']==1)]['FEX_C18'].sum()
print(f'Hombres Desempleados en risaralda: {HDR}')

# 2. Calcular cantidad de hombres en la PEA de risaralda
PEAHR = df_ft[(df_ft['FT']==1) & (df_ft['DPTO']==66) & (df_ft['P3271']==1) ]['FEX_C18'].sum()
print(f'Poblacion Economicamente Activa de hombres en risaralda: {PEAHR}')

# 3. Calcular la tasa de desempleo en hombres de risaralda
TDHR = HDR/PEAHR
print(f'Tasa de Desempleo en hombres de risaralda: {TDHR}')
```

```python
# 1. Calcular cantidad de mujeres desempleados en risaralda
MDR = df_desocupados[(df_desocupados['DSI']==1) & (df_desocupados['DPTO']==66) & (df_desocupados['P3271']==2)]['FEX_C18'].sum()
print(f'mujeres Desempleados en risaralda: {MDR}')

# 2. Calcular cantidad de mujeres en la PEA de risaralda
PEAMR = df_ft[(df_ft['FT']==1) & (df_ft['DPTO']==66) & (df_ft['P3271']==2) ]['FEX_C18'].sum()
print(f'Poblacion Economicamente Activa de mujeres en risaralda: {PEAMR}')

# 3. Calcular la tasa de desempleo en mujeres de risaralda
TDMR = MDR/PEAMR
print(f'Tasa de Desempleo en mujeres de risaralda: {TDMR}')
```

```python
# 1. Calcular cantidad de hombres desempleados en caldas
HDC = df_desocupados[(df_desocupados['DSI']==1) & (df_desocupados['DPTO']==17) & (df_desocupados['P3271']==1)]['FEX_C18'].sum()
print(f'Hombres Desempleados en caldas: {HDC}')

# 2. Calcular cantidad de hombres en la PEA de caldas
PEAHC = df_ft[(df_ft['FT']==1) & (df_ft['DPTO']==17) & (df_ft['P3271']==1) ]['FEX_C18'].sum()
print(f'Poblacion Economicamente Activa de hombres en caldas: {PEAHC}')

# 3. Calcular la tasa de desempleo en hombres de caldas
TDHC = HDC/PEAHC
print(f'Tasa de Desempleo en hombres de caldas: {TDHC}')
```

```python
# 1. Calcular cantidad de mujeres desempleados en caldas
MDC = df_desocupados[(df_desocupados['DSI']==1) & (df_desocupados['DPTO']==17) & (df_desocupados['P3271']==2)]['FEX_C18'].sum()
print(f'Mujeres Desempleados en caldas: {MDC}')

# 2. Calcular cantidad de mujeres en la PEA de caldas
PEAMC = df_ft[(df_ft['FT']==1) & (df_ft['DPTO']==17) & (df_ft['P3271']==2) ]['FEX_C18'].sum()
print(f'Poblacion Economicamente Activa de mujeres en caldas: {PEAMC}')

# 3. Calcular la tasa de desempleo en hombres de caldas
TDMC = MDC/PEAMC
print(f'Tasa de Desempleo en Mujeres de caldas: {TDMC}')
```

```python
# 1. Calcular cantidad de hombres desempleados en tolima
HDT = df_desocupados[(df_desocupados['DSI']==1) & (df_desocupados['DPTO']==73) & (df_desocupados['P3271']==1)]['FEX_C18'].sum()
print(f'Hombres Desempleados en tolima: {HDT}')

# 2. Calcular cantidad de hombres en la PEA de tolima
PEAHT = df_ft[(df_ft['FT']==1) & (df_ft['DPTO']==73) & (df_ft['P3271']==1) ]['FEX_C18'].sum()
print(f'Poblacion Economicamente Activa de hombres en tolima: {PEAHT}')

# 3. Calcular la tasa de desempleo en hombres de tolima
TDHT = HDT/PEAHT
print(f'Tasa de Desempleo en hombres de tolima: {TDHT}')
```
