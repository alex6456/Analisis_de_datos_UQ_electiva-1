```python
import pandas as pd
import numpy as np  
```

```python
#caracteristicas generales, seguridad social en salud y educación 

#links a los conjuntos de datos: 
url_cgsse = 'https://raw.githubusercontent.com/niconomist98/DataAnalyticsUQ/refs/heads/main/Datos/GEIH2025/cgsse.CSV'
url_ocupados = 'https://github.com/niconomist98/DataAnalyticsUQ/raw/refs/heads/main/Datos/GEIH2025/Ocupados.CSV'
url_ft = 'https://github.com/niconomist98/DataAnalyticsUQ/raw/refs/heads/main/Datos/GEIH2025/Fuerza%20de%20trabajo.CSV' 
url_desocupados = 'https://github.com/niconomist98/DataAnalyticsUQ/raw/refs/heads/main/Datos/GEIH2025/No%20ocupados.CSV'
```

```python
# CGSSE
df_cgsse = pd.read_csv(url_cgsse, encoding='latin-1', sep=';', 
                       low_memory=False)

#Ocupados
df_ocupados=pd.read_csv(url_ocupados,encoding='latin-1',sep=';',
                        low_memory=False)

#Fuerza de trabajo
df_ft=pd.read_csv(url_ft,
                  encoding='latin-1',sep=';',low_memory=False)
#No ocupados
df_desocupados=pd.read_csv(url_desocupados,encoding='latin-1',
                           sep=';',low_memory=False)
```

```python
poblacion_total = df_cgsse["FEX_C18"].sum()
print(f"esta es la poblacion total: {poblacion_total}")
```

```python
# codigo de cuantos años tiene es P6040
PET= df_cgsse[(df_cgsse["P6040"]>=15) & (df_cgsse["P6040"]<=65)]["FEX_C18"].sum()
print(f"esta es la poblacion en edad de trabajar es :{PET}")
```

```python
# codigo de cuantos años tiene es P6040
PENT= df_cgsse[(df_cgsse["P6040"]>=00) & (df_cgsse["P6040"]<=15)]["FEX_C18"].sum()
PENT_2= df_cgsse[(df_cgsse["P6040"]>=65)&(df_cgsse["P6040"]<=99)]["FEX_C18"].sum()
print(f"esta es la poblacion que no esta en edad de trabajar es :{PENT + PENT_2}")
```

```python
# pobalcion economicamente activa es variable "FT" en df_ft
# FT=1  si trabaja 
PEA= df_ft[df_ft["FT"]==1]["FEX_C18"].sum()
print(f"poblacion economicamente activa {PEA}")
```
